```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./Output/mentalbert_7labels_with_va_psycholinguistic.csv")

sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="predicted_emotion", y="mean_valence_lex", palette="viridis")
plt.title("Lexical Valence by MentalBERT Predicted Emotion")
plt.xlabel("Predicted Emotion")
plt.ylabel("Lexical Valence (NRC-VAD)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df, 
    x="arousal", 
    y="mean_arousal_lex", 
    alpha=0.3
)
sns.regplot(
    data=df, 
    x="arousal", 
    y="mean_arousal_lex", 
    scatter=False,
    color="red"
)
plt.title("Model Arousal vs Lexical Arousal")
plt.xlabel("Arousal (Circumplex Model)")
plt.ylabel("Lexical Arousal (NRC-VAD)")
plt.show()
```

```python
corr_cols = [
    "mean_valence_lex", "mean_arousal_lex", "mean_dominance_lex",
    "valence_var_lex", "arousal_var_lex",
    "ratio_high_val", "ratio_low_val",
    "ratio_high_arousal", "ratio_low_arousal",
    "anger","disgust","fear","joy","neutral","sadness","surprise",
    "valence","arousal"
]

corr = df[corr_cols].corr()

plt.figure(figsize=(14,10))
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Psycholinguistic + Model Features")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(
    data=df,
    x="subreddit",
    y="mean_valence_lex",
    estimator=np.mean,
    palette="Spectral"
)
plt.xticks(rotation=90)
plt.title("Average Lexical Valence by Subreddit")
plt.ylabel("Mean Valence (Lexicon)")
plt.show()

plt.figure(figsize=(12,6))
sns.barplot(
    data=df,
    x="subreddit",
    y="mean_arousal_lex",
    estimator=np.mean,
    palette="rocket"
)
plt.xticks(rotation=90)
plt.title("Average Lexical Arousal by Subreddit")
plt.ylabel("Mean Arousal (Lexicon)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(
    data=df,
    x="predicted_emotion",
    y="valence_var_lex",
    palette="coolwarm"
)
plt.xticks(rotation=45)
plt.title("Valence Variability by Predicted Emotion")
plt.xlabel("Predicted Emotion")
plt.ylabel("Valence Variance (Lexicon)")
plt.show()

plt.figure(figsize=(12,6))
sns.boxplot(
    data=df,
    x="predicted_emotion",
    y="arousal_var_lex",
    palette="magma"
)
plt.xticks(rotation=45)
plt.title("Arousal Variability by Predicted Emotion")
plt.xlabel("Predicted Emotion")
plt.ylabel("Arousal Variance (Lexicon)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(
    data=df,
    x="predicted_emotion",
    y="ratio_high_arousal",
    estimator=np.mean,
    palette="flare"
)
plt.xticks(rotation=45)
plt.title("High-Arousal Word Ratio by Emotion")
plt.xlabel("Emotion")
plt.ylabel("Ratio of High-Arousal Words")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(
    data=df,
    x="quadrant",
    y="mean_valence_lex",
    palette="viridis"
)
plt.title("Lexical Valence Across Circumplex Quadrants")
plt.xlabel("Quadrant")
plt.ylabel("Mean Lexical Valence")
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(
    data=df,
    x="quadrant",
    y="mean_arousal_lex",
    palette="plasma"
)
plt.title("Lexical Arousal Across Circumplex Quadrants")
plt.xlabel("Quadrant")
plt.ylabel("Mean Lexical Arousal")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(df["mean_valence_lex"], shade=True)
plt.title("Distribution of Lexical Valence")
plt.xlabel("Valence")
plt.show()

plt.figure(figsize=(10,6))
sns.kdeplot(df["mean_arousal_lex"], shade=True, color="red")
plt.title("Distribution of Lexical Arousal")
plt.xlabel("Arousal")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

psy_cols = [
    "mean_valence_lex", "mean_arousal_lex", "mean_dominance_lex",
    "valence_var_lex", "arousal_var_lex",
    "ratio_high_val", "ratio_low_val",
    "ratio_high_arousal", "ratio_low_arousal"
]

X = df[psy_cols].fillna(0)
X_scaled = StandardScaler().fit_transform(X)

pca = PCA(n_components=2)
pca_vals = pca.fit_transform(X_scaled)

df["pca_x"] = pca_vals[:,0]
df["pca_y"] = pca_vals[:,1]

plt.figure(figsize=(10,7))
sns.scatterplot(
    data=df,
    x="pca_x",
    y="pca_y",
    hue="predicted_emotion",
    alpha=0.4,
    palette="tab10"
)
plt.title("PCA of Psycholinguistic Features Coloured by Emotion")
plt.show()
```
